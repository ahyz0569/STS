```python
for i in range(0,90778):
    select_stmt = "INSERT INTO recipe (idx,id,title,size,time,level,description,main,minor,source,img_complete,step,img,ingre_main_oneline) VALUES (%(idx)s,%(id)s,%(title)s,%(size)s,%(time)s,%(level)s,%(description)s,%(main)s,%(minor)s,%(source)s,%(img_complete)s,%(step)s,%(img)s,%(ingre_main_oneline)s)"
    mycursor.execute(select_stmt, {'idx':str(dataset['index'][i]),'id':str(dataset['id'][i]),'title':dataset['title'][i],'size':str(dataset['size'][i]),'time':str(dataset['time'][i]),'level':dataset['level'][i],'description':dataset['description'][i],'main':str(dataset['main_db'][i]),'minor':str(dataset['minor_db'][i]),'source':str(dataset['sauce_db'][i]),'img_complete':dataset['complete'][i],'step':dataset['steps'][i],'img':dataset['steps_img'][i],'ingre_main_oneline':dataset['ingre_main_oneline'][i]})
    print(i)
mydb.commit()
```
